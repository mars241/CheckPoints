```python
# importer les bibliothèques pour le WebScrapping/Parser un site
import requests
from bs4 import BeautifulSoup
```

```python
# Importons la page Web:
# C'est le site de APEC !
# L'Objectif est de Web-Scrapper les fiches de postes de Data Analyst sur le site de l'APEC:
```

```python
URL_data_a = "https://www.apec.fr/tous-nos-metiers/informatique/data-analyst.html"

# Caratéristiques et lien de la page web,puis lancement du "requests"
html = requests.get(URL_data_a)
```

# 1. Requêter l'URL initiale

* **Interroger l'URL initiale**

```python
# Afficher/accèder à la page Web ci-dessus(url)

page_web = html.text
page_web
```

* **Créer soupe**

```python
# Création du "soup" via BeautfulSoup et parsing: c'est-à-dire du nettoyage et de la mise en ordre de ce code HTML

soup = BeautifulSoup(page_web, 'html.parser')
```

# 2. Récupérer les mots clés

```python
# a. les mots clés

# Identifions avec la méthode ".find_all";les objets, classes et des balises et mettons le dans 1 contenaire
items_container = soup.find_all("div", {"class", "card"})
items_container
```

```python
# Voir la longueur du contenaire:
len(items_container)
```

image.png

```python
# REPONSE à : a. les mots clés

# Afficher le bloc "MOTS CLEFS" correspond à balise <div> nommé "rich-text" (Voir image plus haut)

words_key = soup.find("div", class_="rich-text")
words_key
```

```python
# REPONSE à : c. les 2 liens qui permettent d'accéder aux pages de data scientist et data engineer

# Affichez les URL
items_url = soup.find_all("li")
items_url
```

# 3. Récupérer les 4 thèmes d'activites

```python
quatres_themes = items_container[0].find_all("li")
quatres_themes
```

```python
section_themes = soup.find_all('div', {"class" : "rich-text"})
section_themes = section_themes[1].find_all("h3")
section_themes = [st.text.replace("\n", "") for st in section_themes]
section_themes
```

# 4. Récupérer les 2 liens

```python
autres_fiches = soup.find_all("a", {"class" : "small-link"})
autres_fiches
```

```python
liens_fiches = [autre_fiche['href'] for autre_fiche in autres_fiches]
liens_fiches
```

```python
page_liens = soup.find_all("a", {"title": ["Data engineer F/H", "Data scientist F/H"]})
page_liens = [pl.get("href") for pl in page_liens]
page_liens
```
