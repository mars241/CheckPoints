 ## ‚ö° FR - Preprocessing
La colonne "title" contient le domaine, le mill√©sime et le c√©page ("variety").

```
# # #  PREPROCESSING  # # #

# T√©l√©chargement du jeux de donn√©es :
# Dataset des 130k vins : https://github.com/murpi/wilddata/raw/master/wine.zip
# Dataset des 14 vins du Domaine des Croix : https://github.com/murpi/wilddata/raw/master/domaine_des_croix.csv

import pandas as pd
import numpy as np

df_vins = pd.read_csv("https://github.com/murpi/wilddata/raw/master/wine.zip")
df_croix = pd.read_csv("https://github.com/murpi/wilddata/raw/master/domaine_des_croix.csv")
```

```
# Voir le Dataset des 130k vins
df_vins

# RMQ : +sieurs pays ici !
```

```
# Voir le Dataset des 14 vins du Domaine des Croix
df_croix

# RMQ : que la FRANCE concern√© ici !
```

```
# controler les dimensions de 2 DF

print(df_vins.shape)
print(df_croix.shape)

# RMQ: les DF peuvent √™tre merg√©, car il semble avoir les m√™mes colonnes. Pour une question de pratricit√©, on va pouvoir faire une analyse sur un seule DF.
# Quitte √† revenir utiliser chacun des DF pour des besions sp√©cifiques.
```

```
# liste des Variables et leurs types DF all country
df_vins.info()
```

```
# liste des Variables et leurs types DF FRANCE
df_croix.info()
```

```
# Changer le typage de la variable "region_2" de "FLOAT" en "OBJECT' (ou vis versa) dans le DF FRANCE avant le merge
# V√©rification avant des valeurs que contienne cette colonne
df_croix['region_2'].unique()

# RMQ: que des Nan pas de soucis donc, on peut donc merg√© en l'√©tat des choses!
```

```
# Merge/Concat√©nation des 2 DF:
wine = pd.concat([df_vins, df_croix], ignore_index=True)
# voir le DF randomis√© sur 100 enr√©gistrements
wine.sample(100)
```

```
# dimension du DF unifi√©
wine.shape
```

```
# Comme notre DF unififi√© contient DF +sieurs pays & DF FRANCE ,on ose imagin√© qu'il y a certainement des doublons, par pr√©caution on les √©ffacent!
#Nettoyage des doublons
wine = wine.drop_duplicates()
```

```
# traitenemt des NaN/valeur nulles
wine.isna().sum()
```

```
# traitenemt des NaN/valeurs nulles
# 2eme m√©thode:
wine.isnull().sum()

# RMQ : Pour les NaN des pays, d√©signations, 'province', r√©gions 1 & 2, 'taster_name', 'taster_twitter_handle' et 'variety' On ne peut proc√®der √† un remplacement quelconque.
# Par contre , le prix 'price' peut √™tre approfondit (remplacer  les NaN par une moyenne de prix ".mean()" par valeur de d√©signation par exemple .unique() ou mode()/ frequence),
# Mais vu la part des NaN (8406 sur 129971 enr√©gistrements) qui semble √™tre n√©gligeable, on netraitera pas les valeurs nulles
```

```
# REPONSE A LA QUESTION :
# La colonne "title" contient le domaine, le mill√©sime et le c√©page ("variety").
# Tu dois isoler le mill√©sime (l'ann√©e) dans une colonne d√©di√©e.

# Cr√©ation d'une nouvelle colonne dans le DF en extrayant (via les Regex) des dates contenues dans les colonnes title, designation et winery
wine["designation_year"] = wine.designation.str.findall("\d{4}").fillna("0") # cha√Æne de cart√®re des 4 premiers d√©cimals(nombres) excluant 0 dans 'designation'
wine["title_year"] = wine.title.str.findall("\d{4}")
wine["winery_year"] = wine.winery.str.findall("\d{4}")
```

```
# Visualisation de la 1√®re colonne cr√©e:

wine[["designation_year"]]
```

```
# Re-traitement des NaN pour r√©soudre le probl√®me de la visu de l'extraction de la colonne "designation_year"
#Les '0' caract√©risant les nan sont mis dans une liste pour faciliter leur traitement
wine["designation_year"] = wine.designation_year.apply(lambda x: list(x) if x == '0' else x)
```

```
# Re-V√©rification si le probl√®me est r√©solue
wine[["designation_year"]]

# RMQ: Peu d'avanc√©, certaines valeurs restent vides  !
```

```
# Visualisation de la 2eme colonne cr√©e:
wine[["title_year"]]
```

```
# Visualisation de la 3eme colonne cr√©e:
wine[["winery_year"]]
```

```
#Cr√©ation de la colonne mill√©sime qui contient pour l'instant toutes les dates r√©cup√©r√©es
#et filtration des "0"
wine["millesime"] = ((wine.title_year + wine.designation_year + wine.winery_year)
                     .apply(lambda x: [i for i in x if i != "0"]))
```

```
# Afficher la colonne mill√©sime cr√©e
wine[["millesime"]]
```

```
#Drop des dates qui se trouvent plus d'une fois dans la liste
#On part du principe que la date restante est celle du mill√©sime
wine["millesime"] = wine["millesime"].apply(lambda x: [i for i in x if x.count(i) < 2])
```

```
# Utililisation de fonction "def" pour organiser la Base de donn√©es de millesime en forme
def get_millesime(liste_number):
    if len(liste_number) == 1:
        return liste_number[0]
    elif len(liste_number) == 3:
        if int(liste_number[1]) < 2003:
            return max(liste_number)
        else:
            return liste_number[1]
    else:
        return np.nan
```

```
#R√©cup√©ration des valeurs finales et les convertir en FLOAT
wine["millesime extracted"] = wine.millesime.apply(get_millesime).astype(float)
```

```
# Visualisation du mill√©sine affin√© (trait√© & extraite) √† l'aide de Searborn & Plotly
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

wine["millesime extracted"].value_counts().sort_index().plot.bar(figsize=(15,5));
```

```
# Wordcloud= "Nuage de mots"

# importation des biblioth√®ques pour NLP
from wordcloud import WordCloud
import spacy
import nltk.corpus

nlp = spacy.load('en_core_web_sm') # chargement du package 'spacy' pour le nlp


stopwords = nlp.Defaults.stop_words # D√©finition des Stop Words
stopwords.update({"wine", "drink", "fruit", "flavor", "acidity", "finish"}) #rajout de stopwords
wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10) # Param√®trage du Nuage de mots
```

```
#Wordcloud toute la base
sentence = " ".join(wine.sample(1000).description) # Organisation par la jointure du texte contenue dans la colonne'description' en traitement restrint √† une random de 1 000 du DF wine
sent_tokens = nlp(sentence) # Tozennisation des mots contenue dans de DF wine
dico = nltk.FreqDist([token.lemma_ for token in sent_tokens # creation du dictionnaire des tozens(ensemble des mots et des ses tozennis√©s par la m√©thode du "lemmatizer")
                      if (token.lemma_ not in stopwords) and (token.lemma_.isalpha())]) # Alimentation du 'Bag of Word' = "sac de mot"
wordcloud.generate_from_frequencies(dico)
```

```
# R√©presentation du nuage de mot par Mathplotlib

import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)

# RMQ :üëáüèΩ les mots qui ressortent le plus dans les descriptions des vins semblent √™tre: "tannin" ;"cherry" & "palate"
```

```
#Wordcloud Pinot Noir
sentence = " ".join(wine[wine.variety == "Pinot Noir"].sample(1000).description)
sent_tokens = nlp(sentence)
dico = nltk.FreqDist([token.lemma_ for token in sent_tokens
                      if (token.lemma_ not in stopwords) and (token.lemma_.isalpha())])
wordcloud.generate_from_frequencies(dico)

plt.figure(figsize=(10,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)

# RMQ üëáüèΩ: les mots qui ressortent le plus dans les descriptions des vins semblent √™tre:"Cherry"; "Pinot";  "red" ; "raspberry" et "black"
```
