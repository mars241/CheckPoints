```
# REPONSE A LA QUESTION :
# La colonne "title" contient le domaine, le millésime et le cépage ("variety").
# Tu dois isoler le millésime (l'année) dans une colonne dédiée.

# Création d'une nouvelle colonne dans le DF en extrayant (via les Regex) des dates contenues dans les colonnes title, designation et winery
wine["designation_year"] = wine.designation.str.findall("\d{4}").fillna("0") # chaîne de cartère des 4 premiers décimals(nombres) excluant 0 dans 'designation'
wine["title_year"] = wine.title.str.findall("\d{4}")
wine["winery_year"] = wine.winery.str.findall("\d{4}")
```

```
# Visualisation de la 1ère colonne crée:

wine[["designation_year"]]
```

```
# Re-traitement des NaN pour résoudre le problème de la visu de l'extraction de la colonne "designation_year"
#Les '0' caractérisant les nan sont mis dans une liste pour faciliter leur traitement
wine["designation_year"] = wine.designation_year.apply(lambda x: list(x) if x == '0' else x)
```

```
# Re-Vérification si le problème est résolue
wine[["designation_year"]]

# RMQ: Peu d'avancé, certaines valeurs restent vides  !
```

```
# Visualisation de la 2eme colonne crée:
wine[["title_year"]]
```

```
# Visualisation de la 3eme colonne crée:
wine[["winery_year"]]
```

```
#Création de la colonne millésime qui contient pour l'instant toutes les dates récupérées
#et filtration des "0"
wine["millesime"] = ((wine.title_year + wine.designation_year + wine.winery_year)
                     .apply(lambda x: [i for i in x if i != "0"]))
```

```
# Afficher la colonne millésime crée
wine[["millesime"]]
```

```
#Drop des dates qui se trouvent plus d'une fois dans la liste
#On part du principe que la date restante est celle du millésime
wine["millesime"] = wine["millesime"].apply(lambda x: [i for i in x if x.count(i) < 2])
```

```
# Utililisation de fonction "def" pour organiser la Base de données de millesime en forme
def get_millesime(liste_number):
    if len(liste_number) == 1:
        return liste_number[0]
    elif len(liste_number) == 3:
        if int(liste_number[1]) < 2003:
            return max(liste_number)
        else:
            return liste_number[1]
    else:
        return np.nan
```

```
#Récupération des valeurs finales et les convertir en FLOAT
wine["millesime extracted"] = wine.millesime.apply(get_millesime).astype(float)
```

## ☁ 🍱 Descriptions
Quels sont les mots qui ressortent le plus dans les descriptions des vins ? Et spécifiquement pour le pinot noir, est-ce très différent ?  Et pour la province Burgundi en France ?

```
# Wordcloud= "Nuage de mots"

# importation des bibliothèques pour NLP
from wordcloud import WordCloud
import spacy
import nltk.corpus

nlp = spacy.load('en_core_web_sm') # chargement du package 'spacy' pour le nlp


stopwords = nlp.Defaults.stop_words # Définition des Stop Words
stopwords.update({"wine", "drink", "fruit", "flavor", "acidity", "finish"}) #rajout de stopwords
wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10) # Paramètrage du Nuage de mots
```

```
#Wordcloud toute la base
sentence = " ".join(wine.sample(1000).description) # Organisation par la jointure du texte contenue dans la colonne'description' en traitement restrint à une random de 1 000 du DF wine
sent_tokens = nlp(sentence) # Tozennisation des mots contenue dans de DF wine
dico = nltk.FreqDist([token.lemma_ for token in sent_tokens # creation du dictionnaire des tozens(ensemble des mots et des ses tozennisés par la méthode du "lemmatizer")
                      if (token.lemma_ not in stopwords) and (token.lemma_.isalpha())]) # Alimentation du 'Bag of Word' = "sac de mot"
wordcloud.generate_from_frequencies(dico)
```

```
# Répresentation du nuage de mot par Mathplotlib

import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)

# RMQ :👇🏽 les mots qui ressortent le plus dans les descriptions des vins semblent être: "tannin" ;"cherry" & "palate"
```

```
#Wordcloud Pinot Noir
sentence = " ".join(wine[wine.variety == "Pinot Noir"].sample(1000).description)
sent_tokens = nlp(sentence)
dico = nltk.FreqDist([token.lemma_ for token in sent_tokens
                      if (token.lemma_ not in stopwords) and (token.lemma_.isalpha())])
wordcloud.generate_from_frequencies(dico)

plt.figure(figsize=(10,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)

# RMQ 👇🏽: les mots qui ressortent le plus dans les descriptions des vins semblent être:"Cherry"; "Pinot";  "red" ; "raspberry" et "black"
```
