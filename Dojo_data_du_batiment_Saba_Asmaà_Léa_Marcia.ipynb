```python
#pie chart sur rÃ©partition des donnÃ©es
```

```python

data = df.groupby(by='BuildingType')['OSEBuildingID'].nunique()

font_title = {'family': 'serif',
              'color':  '#1d479b',
              'weight': 'bold',
              'size': 18,
             }
```

```python

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(data.values, labels=data.index,
       autopct='%1.1f%%', shadow=True, startangle=30,
       textprops=dict(color="black",size=10, weight="bold"))
ax.axis('equal')
ax.set_title("RÃ©partition des types de bÃ¢timents du Dataset", fontdict=font_title)
plt.show()
```

```python
# Les annÃ©es de construction
```

```python
fig = plt.figure(figsize=(12,8))
ax = sns.histplot(data=data, x='BuildingAge', bins=int((df.DataYear.max() - df.DataYear.min())/5))
ax.set_xlabel("Nombre de bÃ¢timents")
ax.set_ylabel("AnnÃ©e de construction")
plt.title(f"Distribution des annÃ©es de construction des bÃ¢timents\n", fontdict=font_title)
plt.show()
```

```python

fig = plt.figure(figsize=(12,8))
ax = sns.histplot(data=df, x='BuildingAge', bins=int((df.BuildingAge.max() - df.BuildingAge.min())/5))
ax.set_xlabel("Age du bÃ¢timent")
ax.set_ylabel("Nombre de bÃ¢timents")
plt.title(f"Distribution de l'Ã¢ge des bÃ¢timents\n", fontdict=font_title)
plt.show()
```

```python
#nuage de point sur les deux variables de consommation

fig = plt.subplots( figsize=(10,5))
sns.scatterplot(data = df,
                x = 'DataYear',
                y = 'ENERGYSTARScore',
                hue = 'BuildingType')
plt.title('Max de scrore par annÃ©e')
plt.ylabel('Score Maxi')
plt.show()
```

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(corr, annot=True, fmt=".2f", annot_kws={'size':8},
                 mask=mask, center=0, cmap="coolwarm")
plt.title(f"Heatmap des corrÃ©lations linÃ©aires\n",
          fontdict=font_title)
plt.show()
```
