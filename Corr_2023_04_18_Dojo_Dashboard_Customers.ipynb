## Vérifier les valeurs nulles et les doublons

```python
df = pd.read_csv("online_store_customer_data.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Qu'est-ce la colonne Referal ?
df["Referal"].unique()
```

```python
# Incompréhensible en l'état = suppression !
df.drop(['Referal'], axis=1, inplace=True)
df.columns
```

```python
# Supprimer les lignes dont il manque la valeur "Amount_spent" (montant dépensé)
df.dropna(subset=['Amount_spent'], inplace=True)
```

```python
df.isna().sum()
```

```python
# Nombre de lignes dupliquées
df.duplicated().sum()
```

```python
# Affiches les lignes dupliquées
df[df.duplicated(keep=False)].sort_values('Transaction_ID')
```

```python
# Supprimer les doublons
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# Transformer la colonne date en format datetime
df['Transaction_date'] = pd.to_datetime(df['Transaction_date'])
```

```python
df.info()
```

```python
# Transformer la colonne Age en format Int64 (avec une majuscule)
df['Age'] = df['Age'].astype('Int64')
```

***En Pandas, il y a une différence entre int64 et Int64 (avec une majuscule au début).**

int64 fait référence à un type de données entier signé de 64 bits, qui peut contenir des valeurs allant de -2^63 à 2^63-1.*

**Int64 (avec une majuscule au début) est un type de données entier similaire à int64, mais avec la possibilité de représenter des valeurs manquantes (NaN) en plus des valeurs entières.**

En d'autres termes, Int64 est une version avec une optionnalité supplémentaire de int64 qui permet de gérer des valeurs manquantes. Si vous utilisez Int64, vous pouvez stocker des valeurs entières ainsi que des valeurs manquantes dans un tableau Pandas, alors qu'avec int64, les valeurs manquantes ne peuvent pas être stockées.
