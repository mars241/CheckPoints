```python
import pandas as pd
import matplotlib.pyplot as plt



total_spent_name= df.groupby(['State_names'])['Amount_spent'].sum().reset_index()


# Tracer un graphique en barres
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(total_spent_name.index, total_spent_name['Amount_spent'])
ax.set_xticks(total_spent_name.index)
ax.set_xticklabels(total_spent_name['State_names'])
ax.set_xlabel(' tat, Mthode de paiement')
ax.set_ylabel('Montant totales')
ax.set_title('Montant totales en fonction de l\'tat')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Grouper les donnes par ge, tat, mthode de paiement et segment et obtenir la somme des dpenses
total_spent_age = df.groupby(['Age'])['Amount_spent'].sum().reset_index()
total_spent_name = df.groupby(['State_names'])['Amount_spent'].sum().reset_index()
total_spent_segm = df.groupby(['Segment'])['Amount_spent'].sum().reset_index()
total_spent_pay = df.groupby(['Payment_method'])['Amount_spent'].sum().reset_index()

# Tracer un graphique en barres pour chaque figure
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs[0, 0].bar(total_spent_age['Age'], total_spent_age['Amount_spent'])
axs[0, 0].set_title('Dpenses totales en fonction de l\'ge')
axs[0, 1].bar(total_spent_name['State_names'], total_spent_name['Amount_spent'])
axs[0, 1].set_title('Dpenses totales en fonction de l\'tat')
axs[1, 0].bar(total_spent_segm['Segment'], total_spent_segm['Amount_spent'])
axs[1, 0].set_title('Dpenses totales en fonction du segment')
axs[1, 1].bar(total_spent_pay['Payment_method'], total_spent_pay['Amount_spent'])
axs[1, 1].set_title('Dpenses totales en fonction de la mthode de paiement')

# Ajouter des tiquettes pour chaque axe
axs[0, 0].set_xlabel('ge')
axs[0, 0].set_ylabel('Dpenses totales')
axs[0, 1].set_xlabel('tat')
axs[0, 1].set_ylabel('Dpenses totales')
axs[1, 0].set_xlabel('Segment')
axs[1, 0].set_ylabel('Dpenses totales')
axs[1, 1].set_xlabel('Mthode de paiement')
axs[1, 1].set_ylabel('Dpenses totales')

# Ajouter un titre global pour la dataviz
fig.suptitle('Dpenses totales en fonction de l\'ge, de l\'tat, du segment et de la mthode de paiement')

# Ajuster la disposition des sous-graphiques
plt.tight_layout()

# Afficher la dataviz
plt.show()
```
