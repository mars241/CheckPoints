```python
import pandas as pd
import matplotlib.pyplot as plt



total_spent_name= df.groupby(['State_names'])['Amount_spent'].sum().reset_index()


# Tracer un graphique en barres
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(total_spent_name.index, total_spent_name['Amount_spent'])
ax.set_xticks(total_spent_name.index)
ax.set_xticklabels(total_spent_name['State_names'])
ax.set_xlabel(' État, Méthode de paiement')
ax.set_ylabel('Montant totales')
ax.set_title('Montant totales en fonction de l\'état')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Grouper les données par âge, état, méthode de paiement et segment et obtenir la somme des dépenses
total_spent_age = df.groupby(['Age'])['Amount_spent'].sum().reset_index()
total_spent_name = df.groupby(['State_names'])['Amount_spent'].sum().reset_index()
total_spent_segm = df.groupby(['Segment'])['Amount_spent'].sum().reset_index()
total_spent_pay = df.groupby(['Payment_method'])['Amount_spent'].sum().reset_index()

# Tracer un graphique en barres pour chaque figure
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs[0, 0].bar(total_spent_age['Age'], total_spent_age['Amount_spent'])
axs[0, 0].set_title('Dépenses totales en fonction de l\'âge')
axs[0, 1].bar(total_spent_name['State_names'], total_spent_name['Amount_spent'])
axs[0, 1].set_title('Dépenses totales en fonction de l\'état')
axs[1, 0].bar(total_spent_segm['Segment'], total_spent_segm['Amount_spent'])
axs[1, 0].set_title('Dépenses totales en fonction du segment')
axs[1, 1].bar(total_spent_pay['Payment_method'], total_spent_pay['Amount_spent'])
axs[1, 1].set_title('Dépenses totales en fonction de la méthode de paiement')

# Ajouter des étiquettes pour chaque axe
axs[0, 0].set_xlabel('Âge')
axs[0, 0].set_ylabel('Dépenses totales')
axs[0, 1].set_xlabel('État')
axs[0, 1].set_ylabel('Dépenses totales')
axs[1, 0].set_xlabel('Segment')
axs[1, 0].set_ylabel('Dépenses totales')
axs[1, 1].set_xlabel('Méthode de paiement')
axs[1, 1].set_ylabel('Dépenses totales')

# Ajouter un titre global pour la dataviz
fig.suptitle('Dépenses totales en fonction de l\'âge, de l\'état, du segment et de la méthode de paiement')

# Ajuster la disposition des sous-graphiques
plt.tight_layout()

# Afficher la dataviz
plt.show()
```
