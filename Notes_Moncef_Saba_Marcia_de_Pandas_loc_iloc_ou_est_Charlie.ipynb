```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/murpi/wilddata/master/dojos/poke-charlie.csv")
df.head(10)

# For ENGLISH players, they can use this other file for "Where is Waldo?"
# https://raw.githubusercontent.com/murpi/wilddata/master/dojos/poke-waldo.csv
```

```python
df.columns
```

```python
print(type(df[df["Name"].str.contains("Charlie")]["Name"]))
#Rmq1: Mettre "C" Magiscule dans charlie

# le message :<class 'pandas.core.series.Series'> signifie que il maque un argument à appliquer à la fonction sur la DF/ la valeur(A trouver!) ect...
```

```python
#Rmq2: le print() ici sert à afficher le contenue: type ou valeur de la DataFrame
print(df[df["Name"].str.contains("charlie")]["Name"])
```

```python
df[df["Name"].str.contains("Charlie")]
# contient toutes les colonnes qu'on doit retrouver dans la quête pour retouver Charlie!
```

```python
df[df["Name"].str.contains("Charlie")]["Name"][345]
```

```python
#Toutes les lignes dont le le "type 1" est 'Fire'
df[df['Type 1']=='Fire']

# la ligne d'index "279" où la valeur de "Name" est texte.
```

```python
#Charlie est parti. Parmi les Type 1 Fire, il a maintenant l'Attaque maximale
df[df['Type 1']=='Fire'].sort_values(by ='Attack', ascending= False)["Name"]

# la phrase qui nous intéresse est à la 279 index
```

```python
# Maximal <=> ordre croissant:ascending= False (valeur par défaut)
df[df['Type 1']=='Fire'].sort_values(by ='Attack', ascending= False)["Name"][279]
```

```python
#Il s'est encore échappé. Parmi les Rock de type 1, il a maintenant la défense minimum.
df[df['Type 1']== 'Rock'].sort_values(by ='Defense', ascending= True)["Name"]

# la phrase qui nous intéresse est à la 453 index
```

```python
# minimum <=> ordre décroissant :ascending= True
df[df['Type 1']== 'Rock'].sort_values(by ='Defense', ascending= True)["Name"][453]
```
